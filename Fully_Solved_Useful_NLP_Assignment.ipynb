## Question 1

**Compare and contrast NLTK and spaCy in terms of features, ease of use, and performance.**

### Answer

NLTK and spaCy are popular NLP libraries. NLTK is mainly used for education, research, and learning NLP concepts. It provides many algorithms, corpora, and tools for tokenization, stemming, tagging, and parsing. spaCy is designed for production applications and provides fast, efficient processing with pre-trained pipelines.

Comparison:
• Features: NLTK offers extensive NLP resources and datasets; spaCy provides industrial-strength pipelines, NER, dependency parsing, and word vectors.
• Ease of Use: NLTK is flexible but often requires combining multiple modules. spaCy offers simpler APIs and ready-to-use models.
• Performance: spaCy is significantly faster because it is optimized in Cython. NLTK is slower but excellent for learning and experimentation.
• Use Cases: NLTK is preferred in academia and teaching; spaCy is preferred in real-world applications and deployment.

## Question 4

**Describe the architecture and functioning of a Recurrent Neural Network (RNN).**

### Answer

A Recurrent Neural Network (RNN) is designed to process sequential data such as text and speech.

Architecture:
• Input Layer receives sequence elements.
• Hidden Layer maintains a memory state.
• Output Layer generates predictions.

Functioning:
1. Input is processed one element at a time.
2. Hidden state stores information from previous steps.
3. Current output depends on current input and previous hidden state.
4. Weights are shared across time steps.

RNNs capture sequential dependencies but suffer from vanishing gradient problems when handling long sequences.


## Question 5

**What is the key difference between LSTM and GRU networks in NLP applications?**

### Answer

LSTM (Long Short-Term Memory) and GRU (Gated Recurrent Unit) are advanced RNN architectures.

LSTM:
• Uses input, forget, and output gates.
• Maintains a separate cell state.
• More parameters and higher computational cost.
• Better for complex long-term dependencies.

GRU:
• Uses update and reset gates only.
• Combines hidden and cell states.
• Faster and simpler.
• Requires fewer parameters.

GRUs are computationally efficient, while LSTMs often perform better on very long sequences.

## Question 7

### Answer

````
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

text = """Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence."""

tokens = word_tokenize(text)
fdist = FreqDist(tokens)

print(tokens)
print(fdist.most_common(10))

# Sample Output:
# ['Natural', 'Language', 'Processing', ...]
# [('Natural',1), ('Language',1), ...]
````

## Question 8

### Answer

````
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

texts = [
'I love this project',
'This is an amazing experience',
'I hate waiting in line',
'This is the worst service',
'Absolutely fantastic!'
]
labels = [1,1,0,0,1]

tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

seq = tokenizer.texts_to_sequences(texts)
X = pad_sequences(seq, maxlen=5)

model = Sequential([
    Embedding(100, 8, input_length=5),
    LSTM(16),
    Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, labels, epochs=10, verbose=1)

print("Model trained successfully.")
````

## Question 9

### Answer

````
import spacy

nlp = spacy.load('en_core_web_sm')

text = '''Homi Jehangir Bhabha was an Indian nuclear physicist who played a key role in the development of India's atomic energy program.'''

doc = nlp(text)

for token in doc:
    print(token.text, token.lemma_)

print("\nEntities:")
for ent in doc.ents:
    print(ent.text, ent.label_)
````
